```python
github_df = pd.read_csv("github_data.csv")
```

```python
github_df = github_df.drop(columns=['Unnamed: 0'])
```

```python
# Find if website is broken 
def find_website(website_url):
    try:
        r = requests.get(website_url, timeout=3.05)
        website_active = True
    except:
        website_active = False 
    print(website_url)
    return website_active 
```

```python
github_df['website_active'] = github_df['url'].apply(find_website)
```
